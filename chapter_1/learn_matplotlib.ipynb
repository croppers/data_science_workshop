```python
# a convenient tool within matplotlib, which we will study in this module, is pyplot (a.k.a. plt)
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

generating some sample data

```python
x = np.arange(0, 10)
y = 4*np.add(np.sin(x), 0.2*np.random.rand(10)) + 60
```

let's plot this data in a scatter plot

```python
plt.scatter(x, y)
```

we can alter the color and shape of the markers

```python
plt.scatter(x, y, marker = '^', color = 'red')
```

we can add axis and title labels

```python
plt.scatter(x, y, marker = '^', color = 'red')
plt.title('Temperature vs. Time')
plt.ylabel('Temperature')
plt.xlabel('Time')
```

if we change this figure to a plot, then it will draw a line connecting the data

```python
plt.plot(x, y, marker = '^', color = 'red')
plt.title('Temperature vs. Time')
plt.ylabel('Temperature')
plt.xlabel('Time')
```
